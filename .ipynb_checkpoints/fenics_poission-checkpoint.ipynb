# fenics_possion

Runs the test function and plots the results

```python
from fenics_poisson import fenics_poisson_test
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 16
import numpy as np
```

## Input parameters and solver

```python
params = {
    'nx': 31,
    'ny': 31,
    'fe_order': 2,
    'exp_order': 2,
    'length' : 1
}

%timeit u, u_a, L2  = fenics_poisson_test(params['nx'],params['ny'],params['fe_order'],params['exp_order'])
u, u_a, L2  = fenics_poisson_test(params['nx'],params['ny'],params['fe_order'],params['exp_order'])
print('The L2 norm is {:0.8f}'.format(L2))
```

## Plot the results

```python
x_vals = np.linspace(0, params['length'], params['nx'])
y_vals = np.linspace(0, params['length'], params['ny'])

fig,ax = plt.subplots(1,2,figsize=[10,4])
ax[0].pcolor(x_vals,y_vals,u ,vmin = -1 ,vmax = 1);
ax[1].pcolor(x_vals,y_vals,u_a ,vmin = -1 ,vmax = 1);

ax[0].set_ylabel('y')
ax[0].set_xlabel('x')
ax[1].set_xlabel('x')
ax[0].set_title('fd')
ax[1].set_title('an')
```

```python
fig,ax = plt.subplots(1,1,figsize=[5,4])
ax.plot(x_vals,u_a[15,:],linewidth = 2,color = 'b',label = 'an')
ax.plot(x_vals,u[15,:],'.',markersize = 12, color = 'r',label = 'fd')
ax.set_xlim([0,1])
ax.set_ylabel('u')
ax.set_xlabel('x')
fig.legend(frameon = False, loc = [0.7,0.75])
fig.tight_layout()
```
